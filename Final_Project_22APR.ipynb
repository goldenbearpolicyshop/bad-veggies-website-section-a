```python
import requests
from bs4 import BeautifulSoup
import csv

# Define the URL to scrape
url = 'https://courses.my.harvard.edu/psp/courses/EMPLOYEE/EMPL/h/?tab=HU_CLASS_SEARCH&SearchReqJSON=%7B%22ExcludeBracketed%22%3Atrue%2C%22SaveRecent%22%3Afalse%2C%22Facets%22%3A%5B%5D%2C%22PageNumber%22%3A1%2C%22SortOrder%22%3A%5B%22SCORE%22%5D%2C%22TopN%22%3A%22%22%2C%22PageSize%22%3A%22%22%2C%22SearchText%22%3A%22(%20(STRM%3A2232)%20(ACAD_CAREER%3A%5C%22HKS%5C%22)%20(ACAD_ORG%3A%5C%22GOV%5C%22)%20)%20(%20%20)%22%7D'

# Send an HTTP GET request to the URL and retrieve the HTML content
response = requests.get(url)
html_content = response.content

# Use BeautifulSoup to parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table containing the course schedule information
table = soup.find('table', attrs={'class': 'PSLEVEL1GRIDWBO'})

# Extract the column names from the table header
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

# Extract the course schedule information from the table rows
rows = []
for tr in table.find_all('tr')[1:]:
    row = {}
    for i, td in enumerate(tr.find_all('td')):
        row[headers[i]] = td.text.strip()
    rows.append(row)

# Find the links to the course description pages
links = []
for a in table.find_all('a', href=True):
    if 'DERIVED_CLSRCH_DESCRLONG' in a['href']:
        links.append(a['href'])

# Extract the course description information from the description pages
descriptions = []
for link in links:
    response = requests.get(link)
    html_content = response.content
    soup = BeautifulSoup(html_content, 'html.parser')
    description = soup.find('span', attrs={'id': 'DERIVED_CLSRCH_DESCRLONG$0'}).text.strip()
    descriptions.append(description)

# Combine the course schedule and course description information into a list of dictionaries
for i, row in enumerate(rows):
    row['Description'] = descriptions[i]

# Save the data to a CSV file
with open('/content/drive/MyDrive/cs32-class/FP/course_schedule.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=headers + ['Description'])
    writer.writeheader()
    for row in rows:
        writer.writerow(row)

```

```python
import csv
import ipywidgets as widgets
import pandas as pd
import numpy as np

# Open the CSV file and read the data
with open("/content/drive/MyDrive/cs32-class/FP/output.csv", newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    options = [row[1] for row in reader]

print("Choose your courses (use control button - for Windows or command button - for Mac to choose multiple items)")

# Create the dropdown widget
dropdown = widgets.SelectMultiple(options=options)

# Create a button for the user to click
button = widgets.Button(description="Select items")

# Define a function to handle the button click event
def on_button_click(b):
    selected_items = dropdown.value
    print("You have selected the following items:")
    for item in selected_items:
        print(item)
    
    analyze_selected_courses(selected_items)

# Attach the button click handler to the button
button.on_click(on_button_click)

# Display the widgets
display(dropdown)
display(button)

def analyze_selected_courses(selected_courses):
    data = pd.read_csv("/content/drive/MyDrive/cs32-class/FP/output.csv")
    
    # Convert columns to numeric types
    cols = ["overall_class_rating", "overall_faculty_rating", "workload", "instructor_new_insights"]
    data[cols] = data[cols].apply(pd.to_numeric, errors='coerce')

    medians = {
        "overall_class_rating": np.median(data["overall_class_rating"]),
        "overall_faculty_rating": np.median(data["overall_faculty_rating"]),
        "workload": np.median(data["workload"]),
        "instructor_new_insights": np.median(data["instructor_new_insights"]),
    }
    
    selected_data = data[data['class_name'].isin(selected_courses)]
    
    for _, row in selected_data.iterrows():
        compare_and_print_narrative(row["class_name"], row, medians)
    
    averages = selected_data.mean()

    print("\nAverages for the selected classes:")
    for category, avg_value in averages.items():
        if category == "class_name":
            continue
        print(f"The average {category.replace('_', ' ')} is {avg_value:.2f}.")

# Function to compare each class_name data to the median values and print the narrative
def compare_and_print_narrative(class_name, ratings, medians):
    print(f"\nNarrative for {class_name}:")

    for category, median_value in medians.items():
        comparison = (
            "higher" if ratings[category] > median_value else "lower"
        )
        print(
            f"{class_name} has a {comparison} {category.replace('_', ' ')} compared to the overall course data."
        )
```
